```python
import pandas as pd
import numpy as np
import re


df = pd.read_csv(r"C:\Users\MyBook Hype AMD\Downloads\20240121_shopee_sample_data (1).csv")
df.head(10)
```

```python
print(df.info())
print("\nMissing Value:\n", df.isnull().sum())
```

```python
def clean_number(x):
    if pd.isna(x):
        return 0
    
    x = str(x).lower().strip()
    x = x.replace(',', '.')
    
    try:
        if 'k' in x:
            return float(x.replace('k', '')) * 1000
        elif 'rb' in x:
            return float(x.replace('rb', '')) * 1000
        else:
            return float(x)
    except:
        return 0
```

```python
df['favorite'].head(20)
```

```python
def clean_favorite_simple(x):
    if pd.isna(x):
        return 0
    
    x = str(x).lower()
    
    import re
    match = re.search(r'[\d.]+', x)
    if not match:
        return 0
    
    num = float(match.group())
    
    if 'k' in x:
        return num * 1000
    else:
        return num

df['favorite'] = df['favorite'].apply(clean_favorite_simple)
```

```python
df['favorite'].head(20)
```

```python
# numeric utama
df['total_sold'] = df['total_sold'].apply(clean_number)
df['total_rating'] = df['total_rating'].apply(clean_number)

# favorite (format khusus)
df['favorite'] = df['favorite'].apply(clean_favorite_simple)

# rating
df['item_rating'] = pd.to_numeric(df['item_rating'], errors='coerce')

# harga
df['price_actual'] = pd.to_numeric(df['price_actual'], errors='coerce')
df['price_ori'] = pd.to_numeric(df['price_ori'], errors='coerce')
```

```python
# harga saling backup
df['price_ori'] = df['price_ori'].fillna(df['price_actual'])
df['price_actual'] = df['price_actual'].fillna(df['price_ori'])

# isi nilai lain
df['total_sold'] = df['total_sold'].fillna(0)
df['favorite'] = df['favorite'].fillna(0)
df['item_rating'] = df['item_rating'].fillna(0)
df['total_rating'] = df['total_rating'].fillna(0)
```

```python
# revenue
df['revenue'] = df['price_actual'] * df['total_sold']

# discount
df['discount'] = df['price_ori'] - df['price_actual']

# conversion rate (aman dari divide by 0)
df['Sales-to-Favorite Ratio'] = np.where(
    df['favorite'] == 0,
    0,
    df['total_sold'] / df['favorite']
)
```

```python
df = df.drop(columns=[
    'delivery',
    'link_ori',
    'pict_link',
    'idElastic',
    'idHash'
], errors='ignore')
```

```python
print(df[['favorite', 'total_sold', 'Sales-to-Favorite Ratio']].head())

print("\nDeskripsi Favorite:")
print(df['favorite'].describe())

print("\nMissing Value Akhir:")
print(df.isnull().sum())
```

```python
# cek dulu jumlah NaN
print(df[['price_ori', 'price_actual', 'revenue', 'discount']].isnull().sum())

# karena baris bermasalah cuma sedikit, buang baris yang harga masih kosong
df = df.dropna(subset=['price_ori', 'price_actual'])

# hitung ulang revenue dan discount setelah drop NaN
df['revenue'] = df['price_actual'] * df['total_sold']
df['discount'] = df['price_ori'] - df['price_actual']

# kalau ada inf/NaN sisa, amankan
df['revenue'] = df['revenue'].replace([np.inf, -np.inf], 0).fillna(0)
df['discount'] = df['discount'].replace([np.inf, -np.inf], 0).fillna(0)

# rapikan tipe data
df['total_sold'] = df['total_sold'].round(0).astype(int)
df['favorite'] = df['favorite'].round(0).astype(int)
df['total_rating'] = df['total_rating'].round(0).astype(int)
df['revenue'] = df['revenue'].round(0).astype(int)
df['discount'] = df['discount'].round(0).astype(int)

print(df[['price_ori', 'price_actual', 'total_sold', 'favorite', 'revenue', 'discount']].head())
```

```python
# ======================
# 8. EXPORT
# ======================
df.to_csv(r"C:\Users\MyBook Hype AMD\Downloads\shopee_clean.csv", index=False)

print("\nData berhasil dibersihkan dan disimpan sebagai 'shopee_clean.csv'")
```

```python
# total keseluruhan
total_sold_sum = df['total_sold'].sum()
favorite_sum = df['favorite'].sum()

purchase_ratio = total_sold_sum / favorite_sum if favorite_sum != 0 else 0

print("=== TOTAL ===")
print("Total Sold      :", total_sold_sum)
print("Total Favorite  :", favorite_sum)

print("\n=== PURCHASE RATIO ===")
print("Ratio (raw)     :", purchase_ratio)
print("Ratio (%)       : {:.2f}%".format(purchase_ratio * 100))
```

```python
# total keseluruhan
total_sold_sum = df['total_sold'].sum()
total_rating_sum = df['total_rating'].sum()

review_rate = total_rating_sum / total_sold_sum if total_sold_sum != 0 else 0

print("=== TOTAL ===")
print("Total Sold     :", total_sold_sum)
print("Total Rating   :", total_rating_sum)

print("\n=== REVIEW RATE ===")
print("Rate (raw)     :", review_rate)
print("Rate (%)       : {:.2f}%".format(review_rate * 100))
```

```python
# hitung per baris
df['row_review_rate'] = df.apply(
    lambda x: x['total_rating'] / x['total_sold'] if x['total_sold'] != 0 else 0,
    axis=1
)

print("=== RATA-RATA PER PRODUK ===")
print("Mean (raw):", df['row_review_rate'].mean())
print("Mean (%): {:.2f}%".format(df['row_review_rate'].mean() * 100))
```

```python
ratio_sum = total_rating_sum / total_sold_sum
ratio_avg = df['row_review_rate'].mean()

print("=== PERBANDINGAN ===")
print("SUM method (benar)  : {:.2f}%".format(ratio_sum * 100))
print("AVG method (bias)   : {:.2f}%".format(ratio_avg * 100))
```
